```python
import pandas as pd
from scipy.stats import skew, mode

# Charger le fichier CSV
url = "https://github.com/kflisikowski/ds/blob/master/dane_x.csv?raw=true"
mydata = pd.read_csv(url, sep=";", decimal=",")

X = mydata["X"]  # vérifie le nom exact de la colonne !

# Moyenne et écart type
mean_val = X.mean()
std_dev = X.std()

# Skewness
skew_val = skew(X)

# Asymétrie textuelle
skew_type = (
    "Right skew" if skew_val > 0 else
    "Left skew" if skew_val < 0 else
    "Symmetric"
)

# Kurtosis si besoin :
# from scipy.stats import kurtosis
# kurt = kurtosis(X, fisher=False)
# flatness = "less flattened" if kurt > 3 else "flatter"

# Minimum
min_val = X.min()

# 28% quantiles
quant_28 = X.quantile(0.28)
quant_72 = X.quantile(0.72)

# IQR
iqr = X.quantile(0.75) - X.quantile(0.25)

# Médiane
median_val = X.median()

# Mode (valeur la plus fréquente)
mode_val = mode(X, keepdims=False).mode

# Résumé
print(f"Mean: {mean_val:.4f}")
print(f"Standard deviation: {std_dev:.4f}")
print(f"Skewness: {skew_val:.4f} → {skew_type}")
print(f"Min: {min_val}")
print(f"28% at most: {quant_28:.4f}")
print(f"28% at least: {quant_72:.4f}")
print(f"IQR: {iqr:.4f}")
print(f"Median: {median_val:.4f}")
print(f"Mode: {mode_val}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import skew

# Charger le fichier CSV local
df = pd.read_csv("boxhist.csv")  # adapte si le nom est différent
x = df["x"]  # adapte si le nom de colonne est différent

# 1. Histogramme
plt.figure()
plt.hist(x, bins='auto', edgecolor='black')
plt.title("Histogram")
plt.xlabel("x")
plt.ylabel("Frequency")
plt.show()

# 2. Boxplot
plt.figure()
sns.boxplot(x=x)
plt.title("Boxplot")
plt.show()

# 3. Stripchart
plt.figure()
sns.stripplot(x=x, jitter=0.3, size=6)
plt.title("Stripchart")
plt.show()

# 4. Quartiles et médiane
q1 = x.quantile(0.25)
q3 = x.quantile(0.75)
median = x.median()

# 5. Skewness
skewness = skew(x)

# 6. Résumé
print(f"Q1 (25% des obs <)     : {q1:.2f}")
print(f"Q3 (25% des obs >)     : {q3:.2f}")
print(f"Médiane (50% des obs <): {median:.2f}")
print(f"Skewness               : {skewness:.2f}")
```
